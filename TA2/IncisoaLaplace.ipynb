Considerando:  $\frac{\partial ^2u}{\partial x^2}+\frac{\partial ^2u}{\partial y^2}=0$

Con condiciones de frontera en:

0<x<1 u(0,y)=0=u(1,y)

0<x<1 u(x,0)=0 u(x,1)=100 u(x,1)=200

Dada en terminos de la funcion:

$u(x,t)=\sum_{n=1}^{\infty} [200\frac{(1-(-1)^n)}{nπ}cosh(nπ)+200\frac{(1-(-1)^n)(2-cosh(nπ))}{senh(nπ)(nπ)^2}]sen(xnπ)$



```python
from pylab import *
from matplotlib import animation,rc
from IPython.display import HTML 
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
x= linspace(0,1,1000)
y= linspace(0,1,1000)
```

Considerando alserie de Fourier en sus valores 

```python
def u_x_y(x,y):
    N = 10
  
    f = zeros ((len(x),len(y)))
    for w in range(len(y)):
        for n in range(1,N):
          a = (200*(1+(-1)^(n+1)))/(n*pi)
          b = (200*(1+(-1)^(n+1))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,w] += sin(n*pi*x)*(a*cosh(n*pi*y[w])+b*sinh(n*pi*y[w]))
    return f
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="Paired_r"); #Otros colores:"viridis", "inferno", "plasma"
ax.plot_surface(X, Y, U, cmap="Paired_r"); #Otros colores:"viridis", "inferno", "plasma"
```
